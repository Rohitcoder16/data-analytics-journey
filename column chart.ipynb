```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/ACER/Downloads/titanic_data.csv")
```

```python
dataset.head(3)
```

```python
mn=np.mean(dataset["Age"])
```

```python
 mp=np.median(dataset["Age"])
```

```python
mo=dataset["Fare"].mode()[0]
```

```python
sns.histplot(x="Age",data=dataset,bins=[i for i in range(0,67,10)])
plt.plot([mn for i in range(0,50)],[i for i in range(0,50)],c="green",label="mean")
plt.plot([mp for i in range(0,50)],[i for i in range(0,50)],c="blue",label="median")
plt.plot([mo for i in range(0,50)],[i for i in range(0,50)],c="red",label="mode")
plt.legend()
plt.show() 

```
